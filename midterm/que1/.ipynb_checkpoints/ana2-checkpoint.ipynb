```python
def toEmail(file):
    zoo = open(file, 'r',errors='ignore') #had issues with the open function so I forced the open to ignore errors
    moo = zoo.read() #read in file
    email = Parser().parsestr(moo)
    return(email)
```

```python
def sentEmails(name):
    first = name.lower().split(' ')[0]
    last = name.lower().split(' ')[1]
    fName = last+'-'+list(first)[0]
    directory = home+'/midterm/data/enron/maildir'
    dirty = os.path.join(directory,fName)
    mailAdd = first+'.'+last+'@enron.com'
    zoo = dict()
    folds = glob.glob(dirty+'/*')
    for each in folds:
        fids = glob.glob(each+'/*.')
        for n in fids:
            email = toEmail(n)
            if email['from'] == mailAdd:
                sent = email['to']
                if sent is not None:
                    if ',' in sent:
                        sent = sent.split(',')
                        sent = [s.strip() for s in sent]
                        for x in sent:
                            if x in zoo:
                                zoo[x] += 1
                            else:
                                zoo[x] = 1
                    else:
                        if sent in zoo:
                            zoo[sent] += 1
                        else:
                            zoo[sent] = 1
    return zoo
```

```python
def emailsFromXtoX(sender,receiver):
    sendFirst = sender.lower().split(' ')[0]
    sendLast = sender.lower().split(' ')[1]
    recFirst = receiver.lower().split(' ')[0]
    recLast = receiver.lower().split(' ')[1]
    aList = sentEmails(sendLast+'-'+list(sendFirst)[0])
    bList = receivedEmails(recLast+'-'+list(recFirst)[0])
    senderMail = sendFirst+'.'+sendLast+'@enron.com'
    recMail = recFirst+'.'+recLast+'@enron.com'
    if recMail in aList.keys():
        print(sender,'sent',receiver,aList[recMail],'emails')
    if senderMail in bList.keys():
        print(receiver, 'received', bList[senderMail], 'emails from',sender)
    
```

```python
emailsFromXtoX('Kenneth Lay','Jeff Skilling')
emailsFromXtoX('Jeff Skilling','Kenneth Lay')
```

```python
emailsFromXtoX('Kenneth Lay','Andrew Fastow')
```

```python
def emailsFromXtoXwithFiles(sender,receiver):
    sendFirst = sender.lower().split(' ')[0]
    sendLast = sender.lower().split(' ')[1]
    recFirst = receiver.lower().split(' ')[0]
    recLast = receiver.lower().split(' ')[1]
    aList = sentByName(sendLast+'-'+list(sendFirst)[0])
    bList = receivedByName(recLast+'-'+list(recFirst)[0])
    senderMail = sendFirst+'.'+sendLast+'@enron.com'
    recMail = recFirst+'.'+recLast+'@enron.com'
    if recMail in aList.keys():
        print(sender,'sent',receiver,len(aList[recMail]),'emails')
    if senderMail in bList.keys():
        print(receiver, 'received', len(bList[senderMail]), 'emails from',sender)
    finList = aList[recMail]
    finList.update(bList[senderMail])
    print(len(finList),'total')
```

```python
emailsFromXtoXwithFiles('Kenneth Lay','Jeff Skilling')
```

```python
def emailsBetweenEnron(a,b):
    aFirst = a.lower().split(' ')[0]
    aLast = a.lower().split(' ')[1]
    bFirst = b.lower().split(' ')[0]
    bLast = b.lower().split(' ')[1]
    aMail = aFirst+'.'+aLast+'@enron.com'
    bMail = bFirst+'.'+bLast+'@enron.com'
    aList = sentByName(a) #sent by a
    bList = receivedByName(b) #received by b
    cList = receivedByName(a) #received by a
    dList = sentByName(b) #sent by b
    finList = {}
    if bMail in aList.keys():
        finList.update(aList[bMail])
    if aMail in bList.keys():
        finList.update(bList[aMail])
    if bMail in cList.keys():
        finList.update(cList[bMail])
    if aMail in dList.keys():
        finList.update(dList[aMail])
    print(len(finList),'total')
```

```python
emailsBetweenEnron('Kenneth Lay','Jeff Skilling')
```

```python
emailsBetweenEnron('Kenneth Lay','Andrew Fastow')
emailsBetweenEnron('Andrew Fastow','Jeff Skilling')
```
