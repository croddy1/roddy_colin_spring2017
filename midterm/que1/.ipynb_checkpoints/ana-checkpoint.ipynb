```python
import nltk,glob,os,sys,time,IPython
from email.parser import Parser
from os.path import expanduser
from datetime import datetime
from datetime import date
home = expanduser("~")
sound_file = '/Users/Colin/Documents/horse.wav'
```

```python
directory = home+'/midterm/data/enron/maildir'
peopledir = glob.glob(directory+'/*')
#print(peopledir[0])
```

```python
def toEmail(file):
    zoo = open(file, 'r',errors='ignore') #had issues with the open function so I forced the open to ignore errors
    moo = zoo.read() #read in file
    email = Parser().parsestr(moo)
    return(email)
```

```python
def sentTo(name):
    dirty = directory+'/'+name
    zoo = dict()
    folds = glob.glob(dirty+'/*')
    for each in folds:
        fids = glob.glob(each+'/*.')
        for n in fids:
            email = toEmail(n)
            sent = email['to']
            if sent is not None:
                if ',' in sent:
                    sent = sent.split(',')
                    sent = [s.strip() for s in sent]
                    for x in sent:
                        if x in zoo:
                            zoo[x] += 1
                        else:
                            zoo[x] = 1
                else:
                    if sent in zoo:
                        zoo[sent] += 1
                    else:
                        zoo[sent] = 1
    return zoo
```

```python
def sentFrom(name):
    dirty = directory+'/'+name
    zoo = dict()
    folds = glob.glob(dirty+'/*')
    for each in folds:
        fids = glob.glob(each+'/*.')
        for n in fids:
            email = toEmail(n)
            sent = email['from']
            if sent is not None:
                if ',' in sent:
                    sent = sent.split(',')
                    sent = [s.strip() for s in sent]
                    for x in sent:
                        if x in zoo:
                            zoo[x] += 1
                        else:
                            zoo[x] = 1
                else:
                    if sent in zoo:
                        zoo[sent] += 1
                    else:
                        zoo[sent] = 1
    return zoo
```

```python
def eDate(mail):
    strDate = mail['date'].split('-')[0].split(',')[1]
    date = datetime.strptime(strDate,' %d %B %Y %H:%M:%S ')
    date = datetime.date(date)
    return date
```

```python
def isBetween(dLow, dHigh, email):
    return dLow <= eDate(email) <= dHigh
```

```python
aList = dict()
search = 'Andersen'
start_time = time.time()
for person in peopledir:
    folders = glob.glob(person+'/*')
    name = person.split('/')[-1]
    for folder in folders:
        files = glob.glob(folder+'/*.')
        for file in files:
            email = toEmail(file)
            body = email.get_payload()
            if search in body.lower():
                aList[name] = file
print(len(aList))
print("--- %s seconds ---" % (time.time() - start_time))
IPython.display.Audio(sound_file, autoplay=True)
```
