```python
#FOR TOP STORIES
def topStories(year):
    url = "https://api.nytimes.com/svc/archive/v1/%s/1.json" %year
    r = requests.get(url, headers = apikey)
    today = datetime.date.today()
    date = today.isoformat()
    text = r.json()
    results = text['results']
    for each in results:
        title = each['title']
        section = each['section']
        subsection = each['subsection']

        if subsection is not None:
            directory = home+'/midterm/data/NYT/%s/%s/%s/%s' %(date,subject,section,subsection)
        else:
            directory = home+'/midterm/data/NYT/%s/%s/%s' %(date,subject,section)

        if not os.path.exists(directory):os.makedirs(directory) #create directory if it does not alredy exist
        with open(os.path.join(directory,title),'w+') as outPut:
                json.dump(each,outPut)
```

```python
#FOR ARCHIVE SEARCH
def archiveSearch(year):
    url = "https://api.nytimes.com/svc/archive/v1/%s/1.json" %year
    r = requests.get(url, headers = apikey)
    text = r.json()
    results = text['response']['docs']
    i=0
    for each in results:
        title = each['headline']['main']
        section = each['section_name']
        subsection = each['subsection_name']
        #print(section,subsection)

        if section is not None:
            if subsection is not None:
                directory = home+'/midterm/data/NYT/archive/%s/%s/%s' %(year,section,subsection)
            else:
#                 if ';' in section:
#                     holder = section.split(';')
#                     holder = [x.strip() for x in holder]
#                     section = '/'.join(word for word in holder)
                if len(section.split('/')) > 1:
                    directory = home+'/midterm/data/NYT/archive/%s/%s' %(year,section)
                else:
                    directory = home+'/midterm/data/NYT/archive/%s/%s/No_Subsection' %(year,section)
        else:
            directory = home+'/midterm/data/NYT/archive/%s/No_Section' %year

        if not os.path.exists(directory):os.makedirs(directory) #create directory if it does not alredy exist
        if '/' in title: 
            title = title.replace('/','.')
        try:
            with open(os.path.join(directory,title),'w+') as outPut:
                json.dump(each,outPut)
        except OSError as e:
            #print(os.path.join(directory,title))
            print('---',i)
        i += 1

    #IPython.display.Audio(sound_file, autoplay=True)
```

```python
#Get file paths from folder
def pathsFromFolder(aPath):
    aList = []
    for path, subdirs, files in os.walk(aPath):
        for name in files:
            if name != '.DS_Store':
                aList.append(os.path.join(path,name))
                #print(os.path.join(path, name))
    return aList

#Get file names from folder
def namesFromFolder(aPath):
    aList = []
    for path, subdirs, files in os.walk(aPath):
        for name in files:
            if name != '.DS_Store':
                aList.append(name)
                #print(os.path.join(path, name))
    return aList
```
