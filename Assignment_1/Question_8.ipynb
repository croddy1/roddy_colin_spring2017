- If you find a word in nps_chat but not in wordlist, it is considered an unusual word. (e.g. Heyyyy, gr8, 4ever can be considered unusual words for this analysis)
- Find out how many unusual words people speak in chat (notice: usernames should not be counted)

```python
import nltk
from nltk.corpus import nps_chat 
allWords = nltk.corpus.words.words()
npsWords = nps_chat.words()
```

```python
print(len(allWords))
print(len(npsWords))
zoo = list(set(npsWords))
print(len(zoo))
boo = [word.lower for word in npsWords]
print(len(boo))
coo = list(set(boo))
print(len(coo))

uniqueNpsWords = list(set([word.lower for word in npsWords]))
print(len(set(uniqueNpsWords)))
```

```python
unusual= [word for word in uniqueNpsWords if word not in allWords]
print(len(unusual))
unusual2= [word for word in zoo if word not in allWords]
print(len(unusual2))
```
